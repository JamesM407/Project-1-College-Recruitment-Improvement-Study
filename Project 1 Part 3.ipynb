### Purpose: Analyze the data to answer questions about student enrollment in computing majors using plots

Importing libraries and the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
results = pd.read_csv("Cleaned Majors Survey Results - Fall.csv")
```

Filtering the data to only include computing majors

```python
computing = results[results['degree'] == "computing"]
```

## 1. What factors have the highest impact on students' decisions to enroll at CCM? 

Searches for the columns about the factors that influenced enrollment and then sorts them by the factors that have the most high impacts on students

```python
impact_columns = [col for col in results.columns if col.startswith("impact_")]
impact_counts = results[impact_columns].apply(lambda x: (x == 'High Impact').sum()).sort_values(ascending=False)
```

Plotting the data on a bar graph to make it easier to understand what has the biggest impacts on a student's decision to enroll at CCM

```python
plt.figure(figsize=(12, 6))
plt.barh(impact_counts.index, impact_counts.values, color="blue")
plt.xlabel("Count of Students Indicating High Impact")
plt.ylabel("Factors")
plt.title("Factors with the Highest Impact on Enrollment")
plt.gca().invert_yaxis()
plt.show()
```

The biggest impacts are the ability to transfer CCM credits to a 4 year school, affordable cost, and location convenience

## 2. Which pre-college activities have the highest conversion rates to enrollment? 

Searches for the columns about pre-college activities that students can participate in and then sorts them by the activities that the most student participated in

```python
activity_columns = [col for col in results.columns if col.startswith("particpate_")]
activity_counts = results[activity_columns].apply(lambda x: (x == 'Yes').sum()).sort_values(ascending=False)
```

Calculating the conversion rates

```python
global_enrollment_rate = len(results)
conversion_rates = (activity_counts / global_enrollment_rate * 100).sort_values(ascending=False)
```

Plotting the data on a bar graph to make it easier to understand what has the highest conversion rates to enrollment

```python
plt.figure(figsize=(12, 6))
plt.barh(conversion_rates.index, conversion_rates.values, color="green")
plt.xlabel("Conversion Rate (%)")
plt.ylabel("Pre-College Activities")
plt.title("Pre-College Activities with the Highest Conversion to Enrollment")
plt.gca().invert_yaxis()
plt.show()
```

The biggest conversions to enrollment are individual campus visits, open houses, and  on-campus information sessions

## 3. What are the most influential sources of information for students who enroll in computing courses?

Searches for the columns about the sources of information used by students that enrolled in computing classes and then sorts them by the sources that the most students used

```python
info_columns = [col for col in results.columns if col.startswith("info_")]
info_counts = computing[info_columns].apply(lambda x: (x == 'Yes').sum()).sort_values(ascending=False)
```

Plotting the data on a bar graph to make it easier to understand what the most influential sources of information are

```python
plt.figure(figsize=(12, 6))
plt.barh(info_counts.index, info_counts.values, color="purple")
plt.xlabel("Count of Computing Students Indicating High Influence")
plt.ylabel("Information Sources")
plt.title("Most Influential Information Sources for Computing Students")
plt.gca().invert_yaxis()
plt.show()
```

The most influential sources of information are CCM's admissions, CCM's website, and high school guidance counselors

## 4. What are the major motivations why computing majors seek a computing degree/certificate?

Searches for the columns about the motivations that influenced enrollment and then sorts them by the motivations that the most students had

```python
motivation_columns = [col for col in results.columns if col.startswith("motivation_")]
motivation_counts = computing[motivation_columns].apply(lambda x: (x == 'Yes').sum()).sort_values(ascending=False)
```

Plotting the data on a bar graph to make it easier to understand what motivated students the most

```python
plt.figure(figsize=(12, 6))
plt.barh(motivation_counts.index, motivation_counts.values, color="orange")
plt.xlabel("Count of Computing Majors")
plt.ylabel("Motivations")
plt.title("Top Motivations for Pursuing a Computing Degree")
plt.gca().invert_yaxis()
plt.show()
```

The biggest motivations are getting a job in the computing field, transferring to bachelor's level program, and personal enrichment
