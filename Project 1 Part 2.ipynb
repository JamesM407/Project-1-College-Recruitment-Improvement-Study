Condenseing the majors column

```python
def clean_major(major):
    if pd.isna(major):
        return np.nan
    major = str(major).lower()
    if "computer" in major or "software" in major or "information" in major or "cybersecurity" in major:
        return "computing"
    elif "engineer" in major:
        return "engineering"
    elif "business" in major or "management" in major or "finance" in major or "accounting" in major:
        return "business"
    elif "science" in major or "biology" in major or "chemistry" in major or "physics" in major or "mathematics" in major:
        return "science"
    elif "art" in major or "design" in major or "music" in major:
        return "arts"
    elif "health" in major or "nursing" in major or "medical" in major:
        return "health"
    elif "criminal" in major or "justice" in major or "law" in major:
        return "criminal justice"
    elif "education" in major or "teaching" in major:
        return "education"
    elif "liberal" in major or "humanities" in major or "social science" in major:
        return "liberal arts"
    else:
        return "other"

results["degree"] = results["degree"].apply(clean_major)
```
